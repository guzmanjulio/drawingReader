```python
def get_spectrum_colors(names):
  """
  Generates a dictionary of colors across the visual spectrum for a list of names.

  Args:
      names: A list of strings as keys for the color dictionary.

  Returns:
      A dictionary where keys are the names from the input list and values are RGB color tuples.
  """
  # Ensure unique names (avoiding duplicate colors)
  names = list(set(names))

  # Calculate the number of color steps (one for each name)
  num_colors = len(names)
  
  if num_colors < 2:
    step_size = 1
  else:
    step_size = 1.0 / (num_colors) # Normalize step size across the spectrum (0 to 1)

  # Generate hues linearly across the spectrum, starting from red
  hues = [i * step_size for i in range(num_colors)]
  
  print(hues)
  # Convert hues to RGB colors with high saturation and medium brightness
  colors = [colorsys.hsv_to_rgb(h, 1, 1) for h in hues]
  return dict(zip(names, colors))
```

```python
file_path = "IU Mechanical Bulletin 37.pdf"
pdf_document = fitz.open(file_path)
type_of_units = ("BLR","VAV","RVAV","EVAV", "MVAV", "AHU", "FCU","ERU","CP","CHWP","CHWPHZ","HWP","PHP","LEF","IEF","SPF","CEF","KEF","SEF","CRAC","ACURT","CH","FTR","SAV","RAV","EAV","FFTR","CWHEXHZ","HWHEXHZ","PCWHEX","SSF","FFTR","MAU","DHWHEX","HWHEXLZ","RHC")
color_dict = get_spectrum_colors(type_of_units)
unique_patterns = set()
for page_index in range(len(pdf_document)):
        page = pdf_document[page_index]
        print(page)
        bloques = page.get_text("words") 
        i=0
        while i < len(bloques):
            x0, y0, x1, y1, texto, *_ = bloques[i]
            if bloques[i] not in unique_patterns:
                if texto in type_of_units:
                    #print(i,texto)
                    x0_next, y0_next, x1_next, y1_next, texto_next, *_ = bloques[i+1]
                    if y0 != y0_next:
                        texto = texto.rstrip()
                        rect_combined = fitz.Rect(min(x0, x0_next),min(y0, y0_next),max(x1,x1_next),max(y1,y1_next))
                        highlight(rect_combined,texto,texto_next)
                        unique_patterns.add(bloques[i]) 
                        i += 1
                    
            i += 1
            
output_path = file_path[:-4] + " - Mk.pdf"          
pdf_document.save(output_path, garbage=4)
```
